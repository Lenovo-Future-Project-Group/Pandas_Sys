```python
import os
import re
import pandas as pd
```

```python
# 创建总的DataFrame：result_df
# 遍历每个excel文件，读取成dataframe(excel_df)，然后把“日期”、“几楼”、“第几节课”作为列插入到excel_df
# 并将每个excel_ef累加到result_df
def merge_xlsx(xlsx_path, xlsx_name):
    data = pd.DataFrame()
    for file in xlsx_name:
        for path in xlsx_path:
            excel_df = pd.read_excel(path)
            excel_df.rename(columns={'编号': 'id', '名单': 'name', '考勤': 'kq', '说明': 'desc'}, inplace=True)

            # excel_df新增列
            date = get_all_date(file)
            excel_df['date'] = date['data']
            excel_df['room'] = date['room']
            excel_df['class_num'] = date['class_num']

            data = data.append(excel_df)

    return data
```

```python
def get_all_xlsx(file_dir, file_type='.xlsx'):

    path = []
    name = []

    for file in os.listdir(file_dir):
        if os.path.splitext(file)[1] == file_type:
            name.append(file)

    for root, dirs, files in os.walk(file_dir):
        for file in files:
            if os.path.splitext(file)[1] == file_type:
                path.append(os.path.join(root, file))

    # path = [os.path.join(root, file) for root, dirs, files in os.walk(file_dir) for file in files if os.path.splitext(file)[1] == file_type]
    # name = [file for file in os.listdir(file_dir) if os.path.splitext(file)[1] == file_type]

    data = {'file_path': path, 'file_name': name}

    return data
```

```python
def get_all_date(file_name):
    # 使用正则从文件名中解析出‘日期’、‘几楼’、‘第几节课’
    # 构建正则
    search_obj = re.search(r'(.*)日(.*)第(.*)考', file_name)

    data = {'data':search_obj.group(1) + '日', 'room':search_obj.group(2), 'class_num':search_obj.group(3)}

    return data
```

```python
input_xlsx_path = get_all_xlsx(r'../Data/inputData/考勤数据/', '.xlsx')['file_path']
input_xlsx_name = get_all_xlsx(r'../Data/inputData/考勤数据/', '.xlsx')['file_name']

df = merge_xlsx(input_xlsx_path, input_xlsx_name)
```

```python
df.to_csv('../Data/inputData/考勤数据/考勤数据.csv', encoding='utf_8', index=False)
```
